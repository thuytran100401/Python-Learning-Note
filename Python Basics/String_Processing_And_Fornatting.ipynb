```python
# 1. f-Strings (Formatting Output)

name = "Thuy"
drink = "milk tea"
price = 5.5

print(f"{name} ordered a {drink} for ${price}.")
```

```python
 # format number
pi = 3.14159
print(f"Pi rounded to 2 decimals: {pi:.2f}")
```

```python
# 2. Trimming and Cleaning Text
# Useful for cleaning user input or messy data

text = "       Xing Xing Cafe    "
# remove spaces
print(text.strip())

messy = "hello!!!"
# remove '!'
print(messy.strip("!")) 
```

```python
# 3. Search Inside Strings
s = "Welcome to Xing Xing Cafe"
print("Xing" in s)
print(s.startswith("Welcome"))
print(s.endswith("Cafe"))
```

```python
# 4. Split and Join
sentence = "milk tea with boba"

# split a sentence into words 
words = sentence.split()
print(words)

# join them back together
joined = "-".join(words)
print(joined)
```

```python
# practice problem 1: write a function word_count(s) that counts how many words are in a sentence.

def word_count(s):
    # split() break 
    words = s.split()
    return len(words)

# test 
print(word_count("Xing Xing Cafe has tasty milk tea"))
```

```python
# practice problem 2: clean sentence: remove spaces and make lowercase

def clean_sentence(s):
    cleaned = s.replace(" ", "").lower()
    return cleaned

# Test
print(clean_sentence("Xing Xing Cafe has tasty milk tea"))

```
