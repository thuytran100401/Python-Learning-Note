```python
# 1. Dictionaries: A dictionary stores key-value pairs.

# Create a Dictionary
person = {
    "name": "Thuy",
    "age": 25,
    "city": "San Jose"
}

# Access Values
print(person["name"])  # Thuy

# Add / Update Values
person["age"] = 26      # Update
person["job"] = "Engineer"    # Add

# Remove a Key
person.pop("city")

# Loop Through Dictionary
for key, value in person.items():
    print(key, ":", value)
```

```python
# Commom Dictionary Methods

# Create a Dictionary
scores = {"Alice": 90, "Bob": 85, "Carol": 88}

# All Key
print(scores.keys())

# All Values
print(scores.values())

# Key-value pairs
print(scores.items())

# Safely get value
print(scores.get("Alice"))

# Remove key
scores.pop("Bob")

# Count pairs
print(len(scores))
```

```python
# 2. Nested Dictionaries: dictionaries inside dictionaries - used a lot in APIs & JSON.

users = {
    "user1" : {"name": "Alice", "email": "alice@example.com"},
    "user2" : {"name": "Bob", "email" : "bob@example.com"}
}

print(users["user1"]["email"])
```

```python
# 3. Sets 
# Sets are like lists but: no duplicates, no ordered, great for fast lookups

# Create Set
numbers = {1, 2, 3, 3, 4}
print(numbers)

# Add / Remove
numbers.add(5)
numbers.remove(2)
```

```python
# Set Operations
# Union : a | b
# Intersection: a & b
# Difference: a - b
# Example:
a = {1, 2, 3}
b = {3, 4, 5}
print(a & b)
```

```python
# Practice Tasks

# 2. Write a function invert_dict(d)
# Switch keys and values

def invert_dict(d):
    inverted = {}
    for key, value in d.items():
        inverted[value] = key
    return inverted

print(invert_dict({"a": 1, "b": 2}))
```

```python
# 3. Set Practice

# Create two sets
customers_day1 = {"Alice", "Bob", "Thuy"}
customers_day2 = {"Bob", "David", "Eva"}

# Find who came both days
both_days = customers_day1 & customers_day2
print(both_days)

# Find who came only day 1
only_day1 = customers_day1 - customers_day2
print(only_day1)

# Find all unique customers
all_customers = customers_day1 | customers_day2
print(all_customers)
```
