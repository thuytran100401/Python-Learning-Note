```python
# 1. for loops
# Used to repeat a block of code a fixed number of times.

for i in range(5):
    print(i)
    
```

```python
# Problem 1: Sum of N numbers
# Write a function that returns the sum of numbers from 1 to n

def sum_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total

# Test
print(sum_n(5)) #output: 15
```

```python
# Problem 2: Count even numbers in a list 

def count_even(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

# Test
print(count_even([1, 2, 3, 4, 6])) # output: 3
```

```python
# Problem 3: Print the multiplication table of a number 

def multiplication_table(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

# Test 
multiplication_table(3)
              
```

```python
# Problem 4: Write a function factorial(n) that returns the factorial of n.

def factorial(n):
    total = 1
    for i in range(1, n + 1):
        total *= i
    return total

#  test
print(factorial(5))
```
