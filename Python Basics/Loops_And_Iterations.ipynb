```python
# for loops - reqeat a block of code for each item in a sequence (like a list or a range of numbers

for i in range(5):
    print("Hello, Thuy!", i)
```

```python
# loop through a List

fruits = ["apple", "banana", "mago"]
for fruit in fruits:
    print(f"I like {fruit}")
    
```

```python
# using range(start, stop, step)

for num in range(1, 11, 2):
    print(num)
    
```

```python
# practice exercise: print the sum of numbers from 1 to 100

total = 0
for i in range(1, 101):
    total += i
print("Sum =", total)
```

```python
# while loop - runs while a condition is true

count = 1
while count <= 5:
    print("Count is:", count)
    count += 1
    
```

```python
# Example: Guessing Game

secret = 7
guess = 0

while guess != secret:
    guess = int(input("Guess the number (1-10): "))
    if guess < secret:
        print("Too low!")
    elif guess > secret:
        print("Too high!")
    else:
        print("You got it!")
```

```python
# 3. Loop Control - break, continue, pass

# break - stop the loop completely
for num in range(10):
    if num == 5:
        break
    print(num)

# continue - skip one iteration
for num in range(10):
    if num == 5:
        continue
    print(num)

# pass - placeholder (do nothing)
for i in range(5):
    pass 
    
```

```python
# Mini Project: Multiplication Table Generator

number = int(input("Enter a number: "))

print(f"\nMulplication Table for {number}")
for i in range(1, 11):
    print(f"{number} x {i} = {number * i}")
    
```
