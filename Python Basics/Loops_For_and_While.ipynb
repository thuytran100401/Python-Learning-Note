```python
# 1. For Loops
# A for loop repeats a block of code a specific number of times

for i in range(5):
    print("Hello, Xing Xing!")
```

```python
# 2. Using for loop with lists

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)

    
```

```python
# 3. While Loops 
# A while loops repeats as long as a condition is true.

count = 0
while count < 3:
    print("Counting:", count)
    count += 1
```

```python
# Practice Problems 
# 1. write a loop that prints numbers from 1 to 10.

count = 1
while count < 11:
    print("Count:", count)
    count += 1
```

```python
# 2. Write a loop that prints all even numbers from 2 to 20.

count = 0
while count < 21:
    if count % 2 == 0:
        print("Even number:", count)
    count += 1
```

```python
# 3. write a loop that sums numbers from 1 to 100 and prints the result.

count = 1
total = 0
while count < 101:
    total += count
    count += 1

print("Sums from 1 to 100:", total)
```
