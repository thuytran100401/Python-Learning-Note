```python
# 1. List of Dictionaries 
# Example: product list 

products = [
    {"name": "Milk Tea", "price": 5.5, "toppings": ["boba", "jelly"]},
    {"name": "Coffee", "price": 4.0, "toppings": ["cream"]},
    {"name": "Matcha", "price": 6.0, "toppings": ["boba", "cream"]}
]

# Access Date
print(products[0]["name"])
print(products[2]["toppings"])
```

```python
# 2. Loop Through Nested Data

# Print all product names
for product in products:
    print(product["name"])

# Print all toppings for each product
for product in products:
    print(product["name"], "toppings:")
    for topping in product["toppings"]:
        print("-", topping)
```

```python
# 3. Filter Nested Data
# product over $5
expensive = []
for product in products:
    if product["price"] > 5:
        expensive.append(product)

print(expensive)
```

```python
# 4. Dictionary with Lists

orders = {
    "order_1": ["milk tea", "coffee"],
    "order_2": ["matcha", "milk tea"]
}

# Access 
print(orders["order_1"])
```

```python
# 5. Nested Dictionaries

users = {
    "user1": {"name": "Alice", "email": "alice@gmail.com"},
    "user2": {"name": "Bob", "email": "bob@gmail.com"}
}

# Access 
print(users["user1"]["email"])
```

```python
# Practice Tasks
# 1. print all names 

# given list of dictionaries
students = [
    {"name": "Anna", "score": 85},
    {"name": "Ben", "score": 90},
    {"name": "Cara", "score": 78}
]

# print all student names
for student in students:
    print(student["name"])
```

```python
# 2. Find High Scores
# Print names of students with score > 85
for student in students:
    if student["score"] >= 85:
        print(student["name"])
```

```python
# 3. Count total toppings
# Using products example above, count how many times each topping appears.

def count_toppings(products):
    topping_count = {}

    for product in products:
        for topping in product["toppings"]:
            if topping in topping_count:
                topping_count[topping] += 1
            else: 
                topping_count[topping] = 1
    return topping_count

print(count_toppings(products))
```
