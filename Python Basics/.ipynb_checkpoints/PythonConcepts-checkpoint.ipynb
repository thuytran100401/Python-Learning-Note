```python
#variables

name = "Thuy"
age = 25
price = 19.99
is_student = True

print(name, age, price, is_student)
print(type(name), type(age), type(price), type(is_student))
```

```python
# typee conversion
x = "10"
y = int(x)
z = float(x)
print(x, y, z)
```

```python
#strings

greeting = "Hello, Thuy!"
print(greeting.upper())
print(greeting.lower())
print(greeting[0:5]) #slicing
print(len(greeting))
```

```python
# Arithmetic & Input

a = 5
b = 2
print(a + b, a - b, a * b, a / b, a ** b)

# Input example
name = input("what's your name? ")
print("Nice to meet you,", name)
```
