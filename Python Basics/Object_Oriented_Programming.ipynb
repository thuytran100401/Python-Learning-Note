```python
# 1. What is a Class?
# A class is a blueprint.
# Example:
# Class = Drink
# Object = Milk Tea, Coffee, Matcha

# 2. Define your first class
class Drink:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def describe(self):
        print(f"{self.name} costs ${self.price}")
```

```python
# 3. Create Objects (Instances)
milk_tea = Drink("Milk Tea", 5.50)
coffee = Drink("Coffee", 4.00)

milk_tea.describe()
coffee.describe()
```

```python
# 4. What is self?
# self refers to the current object

milk_tea.describe()
Drink.describe(milk_tea)
```

```python
# 5. Add More Attributes
class Drink: 
    def __init__(self, name, price, toppings):
        self.name = name
        self.price = price
        self.toppings = toppings

    def describe(self):
        print(f"{self.name} costs ${self.price}")
        print("Toppings:", ", ".join(self.toppings))
```

```python
# 6. Modify Object State
milk_tea = Drink("Milk Tea", 5.50, ["boba"])

# Objects can change state
milk_tea.toppings.append("jelly")
milk_tea.price += 0.50
```

```python
# Practice Tasks
# 1. Create a Student Class
# Attributes: name, score
# Method: is_passed() -> return true if score >= 60

class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def is_passed(self):
        return self.score >= 60
    
# Test
math = Student("Thuy", 70)
math.is_passed()
```

```python
# 2. Create a BankAccount class
# Attributes: owner, balance
# Methods: deposit(amount), withdraw(amount), display_balance()

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        return self.balance

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            print("Not enough balance to withdraw.")

    def display_balance(self):
        print(f"Balance: ${self.balance}")

# Test
citibank = BankAccount("Thuy", 500)

# Deposit 
citibank.deposit(200)

# Withdraw 
citibank.withdraw(400)

# Display Balance
citibank.display_balance()
```
