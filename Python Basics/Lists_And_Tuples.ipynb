```python
# 1. LISTS (the most important data structure in Python)

# Create a List
fruits = ["apple", "banana", "cherry"]

# Access Items
print(fruits[0])
print(fruits[-1])

# Update Items
fruits[1] = "orange"

# Add Items
fruits.append("mango")

# Remove Items
fruits.remove("apple")

# Loop through a List
for item in fruits:
    print(item)

# Check if Item exists
if "mango" in fruits:
    print("Found it!")
    
```

```python
# 2. List Methods You Must Know

# Create a List
numbers = [5, 3, 8, 1]

# Add to end
numbers.append(9)

# Insert at position
numbers.insert(2, 5)

# Remove by index
numbers.pop(1)

# Remove by value
numbers.remove(5)

# Sort List
numbers.sort()
print(numbers)

# Reserve List
numbers.reverse()
print(numbers)

# Count items
print(len(numbers))
```

```python
# 3. List Comprehensions (advanced but important)

squares = [x*x for x in range(5)]
print(squares)
```

```python
# 4. Tuples (immutable lists)

# Create a Tuple
colors = ("red", "blue", "green")

# Access Like a List
print(colors[1]) # blue

def get_point():
    return (3, 5)

x, y = get_point()
print(x, y)
```

```python
# 2. Write a function get_even_numbers(list)
# Return all even numbers from a list.

def get_even_numbers(list):
    for item in list:
        if  item % 2 == 0:
            print(item)

# Create a list
numbers = [2, 5, 6, 8, 10, 11, 12, 13, 14]
get_even_numbers(numbers)
```
