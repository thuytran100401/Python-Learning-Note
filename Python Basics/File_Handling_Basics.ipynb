```python
# 1. Opening a File
file = open("example.txt", "r")
content = file.read()
file.close()
```

```python
# 2. Best Practice: with open()
# Automatically closes file
# Cleaser
# Safer

with open("example.txt", "r") as file:
    content = file.read()

print(content)
```

```python
# 3. Reading Files

# Read entire file:
with open("example.txt", "r") as file:
    print(file.read())

# Read line by line
with open("example.txt", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# 4. Writing Files ("w")
# "w" overwrites the file.
with open("example.txt", "w") as file:
    file.write("Hello World\n")
```

```python
# 5. Appending Files ("a")
# Adds to the end 
# Safe for logs

with open("example.txt", "a") as f:
    f.write("New line added\n")
```

```python
# 6. Common File Patterns

# Save a list to a file
tasks = ["Buy milk", "Study Python", "Workout"]

with open("example.txt", "w") as f:
    for task in tasks:
        f.write(task + "\n")

# Load list from file
tasks = []
with open("example.txt", "r") as f:
    for line in f:
        tasks.append(line.strip())
```

```python
# 7. Handling Missing Files

try:
    with open("example.txt", "r") as f:
        print(f.read())
except FileNotFoundError:
    print("File not found.")
```

```python
# Practice Tasks
# 1. Write to a file
# Create a program that: Asks user for 3 favorite foods 
# Saves them to foods.txt

favorite_foods = []

for i in range(3):
    food = input(f"Enter favorite food #{i+1}: ")
    favorite_foods.append(food)

with open("foods.txt", "w") as f:
    for food in favorite_foods:
        f.write(food + "\n")

print("Favorite foods saved to foods.txt")

# 2. Read from a file - print each food
with open("foods.txt", "r") as f:
    for line in f:
        print(line.strip())

# 3. Append to a file - add new food
new_food = input("Enter another favorite food: ")

with open("foods.txt", "a") as f:
    f.write(new_food + "/n")

print("Food added succesfully.")
```
