```python
"""
Day 19 - Pandas & EDA 

Goal: explore a dataset, practice filtering, summary statistics, and basic insights

"""

# 1. Import libraries and load dataset

import pandas as pd

# Load dataset
df = pd.read_csv("dataset.csv")

# Quick look 
print(df.head())
print(df.info())
```

```python
# 2. Basic Exploration
# Summary statistics 
print(df.describe())

# Count unique values in a column
print(df['name'].value_counts())
print(df['age'].value_counts())

# Check missing values
print(df.isnull().sum())
```

```python
# 3. Filtering and selecrting data

# People in sales 
sales = df[df['department'] == 'Sales']
print(sales.head())

# Females only 
females = df[df['gender'] == 'Female']
print(females.head())

# Select specific columns
subset = df[['name', 'age', 'gender']]
print(subset.head())
```

```python
# 4. Aggregation and grouping

# Average age by gender
print(df.groupby('gender')['age'].mean())

# Average salary by gender
print(df.groupby('gender')['salary'].mean())
```

```python
# 5. Sorting and adding columns

# Sort by age descending
print(df.sort_values('age', ascending=False).head())

# Create a new columns: Age category
df['AgeCategory'] = pd.cut(df['age'], bins = [0, 12, 18, 60, 100])
print(df[['age', 'AgeCategory']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of ages
plt.hist(df['age'], bins=5)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Count plot of gender
sns.countplot(data=df, x='gender')
plt.title("Gender Count")
plt.show()

# Average age by department
df.groupby('department')['age'].mean().plot(kind='bar')
plt.title("Average Age by Department")
plt.show()

# Average salary by department
df.groupby('department')['salary'].mean().plot(kind='bar')
plt.title("Average Salary by Department")
plt.show()
```
