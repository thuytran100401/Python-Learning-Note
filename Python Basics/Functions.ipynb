```python
# Parameters and Arguments

def greet_user(name):
    print(f"Hello, {name}!")

greet_user("Thuy")
greet_user("Alex")
```

```python
# Return Values

def add(a, b):
    return a + b

result = add(5,3)
print("Sum =", result)
```

```python
# Multiple Parameters and Default Values

def power(base, exponent=2):
    return base ** exponent

print(power(3))
# 3^2 = 9

print(power(3, 4)) 
# 3^4 = 81
```

```python
# Practice Problems

# Write a function square(n) that returns the square of a number,

def square(n):
    return n * n


# Write a function is_even(n) that returns True if the number is even, otherwise False

def is_even(n):
    if n % 2 == 0:
        return True
    else:
        return False


# Write a function max_of_two(a, b) that returns the larger of two numbers.

def max_of_two(a, b):
    if a > b:
        return a
    else:
        return b


print("Square:", square(4))

print("Even:", is_even(3))

print("Max of two:", max_of_two(3,5))
```

```python
# Functions With Loops

def print_table(num):
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

print_table(5)
```
